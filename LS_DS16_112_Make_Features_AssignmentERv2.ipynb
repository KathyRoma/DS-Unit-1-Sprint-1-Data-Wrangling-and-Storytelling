```python
# Let's write a function for this action as long as we will need it to fix the whole column
def term_to_int(term):
  return int(term.strip().strip('months'))
```

```python
term_to_int(term)
# Works nicely
```

```python
# Create a new column with numeric values (so we can manipulate this data mathematically)
new_term = []
for term in df2['term']:
  new_term.append(int(term.strip().strip('months')))

df2['new_term'] = pd.Series(new_term)
```

```python
# The new column is in place
df2.head()
```

```python
# Use .apply method for the sake of time and beauty
## It does the same thing!
df2['term_int'] = df2['term'].apply(term_to_int)
df2.head()
```

```python
# Now we can count averages and medians
df2['new_term'].mean()
```

```python
# Assign a new variable and create an empty list for further manipulations
loan_statuses = ['Current', 'Late (31-120 days)', 'In Grace Period', 'Default', 'Fully Paid']
loan_status_int = []
```

```python
# Let's write a function that identifies if the person is doing great with their money
def status_great(status_string):
  if status_string=='Current' or status_string=='Fully Paid':
   return "Great"
  else:
   return "Not so great"

for status in loan_statuses:
  print(status_great(status))
```

```python
def status_great(status_string):
  if status_string=='Current' or status_string=='Fully Paid':
    return 1
  else:
    return 0
# Use .append method to fill in our empty list

for status in loan_statuses:
  loan_status_int.append(status_great(status))  

loan_status_int
```

```python
df2['Great_status'] = df2['loan_status'].apply(status_great)

df2.head()
# Works smoothly -- just scroll right to see the new column
```
